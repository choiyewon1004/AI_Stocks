# 1.2 이동평균선
- 함수 : rolling(window=요일).mean()
- 오늘의 이동평균선도 계속 변경되며 장이 마감되는 3시 30분에 확정됨

```python
kospiMA5 = kospi['Close'].rolling(window=5).mean() #종가에 대한 5일 이동평균
kospiMA5
```

```python
plt.figure(figsize=(15,5))
plt.plot(kospi["Close"], label="KOSPI")
plt.plot(kospiMA5, label="KOSPIMA5")
plt.legend(fontsize=16)
```
